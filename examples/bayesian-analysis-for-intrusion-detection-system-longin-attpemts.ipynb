# Bayesian Analysis for Intrusion Detection Systems (IDS)

An Intrusion Detection System (IDS) is responsible for monitoring login attempts and identifying potential cyberattacks.

## **System Characteristics**
- The IDS correctly detects actual cyberattacks **90% of the time** (**true positive rate = 90%**).
- The IDS mistakenly flags legitimate user activity as an attack **10% of the time** (**false positive rate = 10%**).
- Statistical data indicates that **only 2% of all login attempts are actual attacks**.

## **Objective**
When an alert is triggered, security teams must determine the likelihood that it represents a **real attack** rather than a **false alarm**.

Using **Bayesâ€™ Theorem**, this implementation calculates the probability that an alert corresponds to a **real cyberattack**.

## **Inputs**
- The **baseline probability** of an attack occurring (`P(Attack)`)
- The **true positive rate** (sensitivity of IDS, `P(Alert | Attack)`)
- The **false positive rate** (false alarms by IDS, `P(Alert | No Attack)`)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define the Given Probabilities
p_attack = 0.02  # 2% probability that a request is an actual attack
p_alert_given_attack = 0.90  # 90% probability of detecting an attack (true positive rate)
p_alert_given_no_attack = 0.10  # 10% probability of a false positive alert

# Compute the Total Probability of an Alert (Law of Total Probability)
p_alert = (p_alert_given_attack * p_attack) + (p_alert_given_no_attack * (1 - p_attack))

# Compute the Probability of an Attack Given an Alert being real (Bayes' Theorem)
p_attack_given_alert = (p_alert_given_attack * p_attack) / p_alert


# Outputs
# Display the Computed Probability
print(f"Probability that an alert is a real attack: {p_attack_given_alert:.4f}")

# Visualize the Bayesian Probability
labels = ["Real Attack", "False Alarm"]
# 
probabilities = [p_attack_given_alert, 1 - p_attack_given_alert]

plt.figure(figsize=(6, 4))
sns.barplot(x=labels, y=probabilities, hue=labels, palette="Purples", edgecolor="black")
plt.ylim(0, 1)
plt.title("Bayesian Probability of an Attack Given an Alert")
plt.ylabel("Probability")
plt.show()
```
